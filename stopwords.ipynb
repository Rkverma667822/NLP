```python
paragraph="""Dr. A.P.J. Abdul Kalam, widely known as the “Missile Man of India,” was one of the most inspiring leaders the world has ever seen. Born on 15 October 1931 in Rameswaram, Tamil Nadu, he came from a humble background, yet he rose to become the 11th President of India and a global symbol of knowledge, discipline, and simplicity.

Kalam grew up in a financially struggling family, but his curiosity and dedication towards studies made him stand out from an early age. He graduated in Physics and later studied Aerospace Engineering at the Madras Institute of Technology. His journey from a small town boy to India’s top scientist shows how determination can overcome every barrier.

As a scientist, Kalam played a major role in India’s defense and space programs. He contributed significantly to the development of India’s first satellite launch vehicle (SLV-III), which placed the Rohini satellite in orbit. His crucial involvement in the development of ballistic missiles like Agni and Prithvi earned him the title “Missile Man.” He was also a key figure in India’s 1998 nuclear tests at Pokhran, making the nation stronger and more confident on the global stage.

In 2002, Kalam became the President of India. Even after reaching such a high position, he remained extremely simple, humble, and connected with the youth. He believed that students are the future of the nation and always encouraged them to dream big. His famous quote, “Dream, dream, dream. Dreams transform into thoughts and thoughts result in action,” continues to inspire millions.

Kalam was not just a scientist or a president—he was a teacher at heart. Even after completing his presidential term, he spent his life teaching, writing books, and motivating young minds. On 27 July 2015, he passed away while delivering a lecture at IIM Shillong, doing what he loved the most.

Dr. A.P.J. Abdul Kalam’s life is a reminder that with hard work, honesty, and courage, anyone can achieve greatness. He remains a role model for students and a true icon of India’s progress."""
```

```python
from nltk.stem import PorterStemmer
from nltk.corpus import stopwords
```

```python
import nltk
nltk.download('stopwords')
```

```python
stopwords.words('English')
```

```python
stemmer=PorterStemmer()
sentences=nltk.sent_tokenize(paragraph)
```

```python
type(sentences)
```

### Apply stopword and filter and then the apply stemming


```python
for i in range(len(sentences)):
    sentences[i]=sentences[i].upper()
    words=nltk.word_tokenize(sentences[i])
    words=[stemmer.stem(word) for word in words if word not in set(stopwords.words('English'))]
    sentences[i]=' '.join(words)
```

```python
from nltk.stem import SnowballStemmer
snowaballs=SnowballStemmer('english')
```

```python
# Apply the stopwords and filter and filter and then the apply snowaballs
for i in range(len(sentences)):
    # sentences[i]
    words=nltk.word_tokenize(sentences[i])
    words=[snowaballs.stem(word) for word in words if word not in set(stopwords.words('English'))]
    sentences[i]=' '.join(words)
```

```python
from nltk.stem import WordNetLemmatizer
lemmittizer=WordNetLemmatizer()
# Apply the stopwords and filter and filter and then the apply lemmittizer
for i in range(len(sentences)):
    # sentences[i]
    words=nltk.word_tokenize(sentences[i])
    words=[lemmittizer.lemmatize(word) for word in words if word not in set(stopwords.words('English'))]
    sentences[i]=' '.join(words)
```
