```python
words=['eat','eaten','goes','going','program','programs','programing','writes','writing']
```

```python
from nltk.stem import PorterStemmer 
stimming=PorterStemmer()
```

```python
for word in words:
    print(word + "....>"+stimming.stem(word))
```

```python
stimming.stem("congratulations")
```

```python
stimming.stem('sitting')
```

```python
stimming.stem('Rajeev')
```

```python
from nltk.stem import RegexpStemmer
regStemmer=RegexpStemmer('ing$|s$|ables$|e',min=4)
```

```python
regStemmer.stem("incoming")
```

```python
regStemmer.stem("goes")
```

```python
regStemmer.stem("fairly")
```

```python
for word in words:
    print(word+"....>"+regStemmer.stem(word))
```
